```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Rhitik9579/Desktop/churn-prediction/data/raw/Telco-Customer-Churn.csv")  # one level up from notebooks folder

df.head()
```

```python
sns.countplot(data=df, x="Churn")
plt.title("Churn vs Non-Churn Count")
plt.show()
```

```python
numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Contract", hue="Churn")
plt.title("Churn Rate by Contract Type")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x="PaymentMethod", hue="Churn")
plt.title("Churn Rate by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

### Key EDA Insights
- Customers with month-to-month contracts churn significantly more.
- Higher churn seen in customers using Electronic Check.
- Tenure is highly correlated with churn — low-tenure customers churn more.
- Customers with higher monthly charges tend to churn more.
- Missing values exist in TotalCharges due to spaces → to be fixed in feature engineering.
